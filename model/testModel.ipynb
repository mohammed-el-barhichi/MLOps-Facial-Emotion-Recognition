```python
import cv2
import numpy as np
import joblib
import os
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
```

```python
# Load the trained model
model = joblib.load("emotion_model.joblib")

```

```python
emotion_dict = {
    0: "angry",
    1: "disgusted",
    2: "fearful",
    3: "happy",
    4: "neutral",
    5: "sad",
    6: "surprised",
}
```

```python
def predict_emotion(image_path):
    # Read the test image
    image = cv2.imread(image_path)

    # Check if the image is empty
    if image is None:
        raise ValueError("Failed to read the image")

    # Preprocess the image
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    roi_gray = cv2.resize(gray_image, (48, 48))
    roi_gray = np.expand_dims(np.expand_dims(roi_gray, -1), 0)

    # Predict the emotion
    emotion_prediction = model.predict(roi_gray)
    max_index = int(np.argmax(emotion_prediction))

    # Get the predicted emotion label
    predicted_emotion = emotion_dict[max_index]

    return predicted_emotion
```

```python
print(predict_emotion("../datas/test/angry/PrivateTest_88305.jpg"))
```

```python

# Calculate and print accuracy
accuracy = np.mean(predict_emotion("../datas/test/angry/PrivateTest_88305.jpg") == 'angry')
print(f"Accuracy: {accuracy}")
```

```python
test_data_folder = '../datas/test'

true_labels = []
predicted_labels = []

for emotion_class in os.listdir(test_data_folder):
    emotion_path = os.path.join(test_data_folder, emotion_class)
    
    if os.path.isdir(emotion_path):

        for image_file in os.listdir(emotion_path):

            true_label = emotion_class 
            true_labels.append(true_label)
            
            image_path = os.path.join(emotion_path, image_file)

            # The rest of your code for predicting emotions
            emotion_prediction = predict_emotion(image_path)

            predicted_label = emotion_prediction
            
            predicted_labels.append(predicted_label)
```

```python
# Convert lists to numpy arrays for compatibility with scikit-learn metrics
true_labels = np.array(true_labels)
predicted_labels = np.array(predicted_labels)

print(true_labels)
print(predicted_labels)
```

```python
# Calculate and print accuracy
accuracy = np.mean(true_labels == predicted_labels)
print(f"Accuracy: {accuracy}")

# Calculate and print classification report
class_report = classification_report(true_labels, predicted_labels)
print("Classification Report:\n", class_report)
```

```python
conf_matrix = confusion_matrix(true_labels, predicted_labels)

plt.imshow(conf_matrix, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()

classes = list(emotion_dict.values())
tick_marks = np.arange(len(classes))
plt.xticks(tick_marks, classes, rotation=45)
plt.yticks(tick_marks, classes)

plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
